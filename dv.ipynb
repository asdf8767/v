```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Line graphs
a=[1,2,3,4,5,6]
b=[10,30,40,20,60,50]
plt.plot(a,b)
plt.xlabel("X value")
plt.ylabel("y value")
plt.title("Heading")
```

```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python
x=df["PassengerId"][:50]
y=df["Fare"][:50]
plt.figure(figsize=(20,8))
plt.plot(x,y)
plt.xlabel("PassengerId")
plt.ylabel("Fare")
plt.title("PassengerId vs Fare")
plt.show()
```

Bar Chart 

```python
cat=["A","B","C","D","E"]
val=[20,30,10,50,100]
plt.bar(cat,val)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Bar graph")
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex",y="Survived", data=df)
plt.title("sex vs Survived")
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.show()

```

```python
embarked_counts = df['Embarked'].value_counts()
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%')
plt.title('Passenger Embarkation Distribution')
plt.axis('equal')
```
